```python
col_names =  ['A']
n  = pd.DataFrame(columns = col_names)
n['y']  = df['VENTAS GOY'] - df['ventas competidor b']
n['b_p']  = df['PRECIO GOY'] - df['PRECIO COMP B']
n['b_i']  = df['inv medios GOY'] - df['INV MEDIOS COMP B']
n['b_d']  = df['DIST MANEJANTES GOY'] - df['DIS MANEJANTES COMP B']
n['b_o']  = df['DIT OFERTAS GOY'] - df['DIS OFERTAS COMP B']
n
```

```python
n['y']  = n['y'].shift(1)- n['y']
n['b_p']  = n['b_p'].shift(1) - n['b_p']
n['b_i']  = n['b_i'].shift(1) - n['b_i']
n['b_d']  = n['b_d'].shift(1) - n['b_d']
n['b_o']  = n['b_o'].shift(1) - n['b_o']
n = n.iloc[1:] 
n
```

```python
import statsmodels.api as sm
endog = n['y']
exog = sm.add_constant(n[['b_p', 'b_o', 'b_d', 'b_i']])

# Fit the model
mod = sm.tsa.statespace.SARIMAX(endog, exog, order=(1,0,0))
fit_res = mod.fit(disp=False)
print(fit_res.summary())
```

```python
mod = sm.tsa.statespace.SARIMAX(endog, exog=exog, order=(0,0,1))
res = mod.filter(fit_res.params)
predict = res.get_prediction()
predict_dy = res.get_prediction(dynamic=22)
predict_dy_ci = predict_dy.conf_int()
predict_ci = predict.conf_int()
```
